```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
df=pd.read_csv('/content/sonar_csv.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(df.drop("Class",axis=1))
```

```python
s=sc.transform(df.drop("Class",axis=1))
```

```python
s
```

collection of columns and rows are called dataframe

```python
a=pd.DataFrame(s,columns=df.columns[:-1])
```

```python
a
```

training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x=a
y=df.Class
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
k=KNeighborsClassifier(n_neighbors=1)
```

```python
k.fit(x_train,y_train)
```

```python
p=k.predict(x_test)
```

```python
p
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,p))
print(classification_report(y_test,p))
```

analyzing better k value through iterations

```python
error_rate=[]
for i in range(1,40):
  k=KNeighborsClassifier(n_neighbors=i)
  k.fit(x_train,y_train)
  p_i=k.predict(x_test)
  error_rate.append(np.mean(p_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color="blue",linestyle="dashed",marker="s",markerfacecolor="orange",markersize=10)
plt.title("Error rate vs k values")
plt.xlabel("k")
plt.ylabel("error rate")
```

```python
k=KNeighborsClassifier(n_neighbors=3)
k.fit(x_train,y_train)
p=k.predict(x_test)
print(confusion_matrix(y_test,p))
print(classification_report(y_test,p))
```
