**MATPLOTLIB:**

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph')
plt.show()
```

```python
plt.subplot(1,2,1)#(row,column,number)
plt.plot(x,y,'b*-')
plt.subplot(1,2,2)
plt.plot(x,y,'r--')
```

```python
plt.subplot(1,3,1)#(row,column,number)
plt.plot(x,y,'b*-')
plt.subplot(1,3,2)
plt.plot(x,y,'r--')
plt.subplot(1,3,3)
plt.plot(x,y,'g')
```

```python
fig=plt.figure()#empty canvas
axes=fig.add_axes([0.1,0.1,1.0,1.0])#left,bottom,width,height(0 to 1)
#plot on that set of axes
axes.plot(x,y,'b*-')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.7,0.3,0.89,0.8])
axes2=fig.add_axes([0.78,0.6,0.4,0.4])
axes1.plot(x,y,'b')
axes2.plot(y,x,'g')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes1.set_title('graph1')
axes2.set_title('graph2')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('graph')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
for ax in axes:
  ax.plot(x,y,'g')
  ax.set_xlabel('x axis')
  ax.set_ylabel('y axis')
  ax.set_title('graph')
  fig.tight_layout()
```

```python
fig
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
```

```python
fig,axes=plt.subplots(figsize=(10,5))
axes.plot(x,y,'r')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('graph')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend(loc=2)#upper left corner
```

```python
ax.legend(loc=1)#upper left corner
ax.legend(loc=3)#bottom left corner
ax.legend(loc=4)#bottom right corner

```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x+1,color="red",linewidth=0.5,ls=':')
ax.plot(x,x+2,color="green",lw=1,linestyle='-')
ax.plot(x,x+3,color="blue",lw=2,marker="+")
ax.plot(x,x+4,color="red",lw=3)
ax.plot(x,x+5,color="blue",lw=3,marker="*")
ax.plot(x,x+6,color="blue",lw=3,marker="o",markersize=8,markerfacecolor="green",markeredgecolor="yellow",markeredgewidth=2)
ax.plot(x,x+7,color="blue",lw=3,marker="s",markersize=6,markerfacecolor="yellow",markeredgecolor="green",markeredgewidth=3)
ax.plot(x,x+8,color="blue",lw=3,marker=",")
```
